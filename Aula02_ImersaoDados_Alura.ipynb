```python
import pandas as pd

# URL da base de dados
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

# Variável que guarda as infomações que serão lidas pelo "pd.read_csv" vindas da base de dados (url_dados) e descompacta da pasta zipada
dados = pd.read_csv(url_dados, compression = 'zip')

#print(dados)
dados # Dataframe
```

```python
# renomear a coluna 'droga' > 'composto' (inplace = True, para gravar a alteração)
mapa = {'droga': 'composto'}
dados.rename(columns = mapa, inplace = True)
```

```python
dados.head()
```

```python
# pegar os 5 primeiros elementos da coluna composto
#dados['composto'].value_counts()[0:5]
compostos_masc = dados['composto'].value_counts()[0:5]
```

```python
# coletar somente o index
comp_index = compostos_masc.index
print(comp_index)
```

```python
# filtrar e selecionar somente as informações que queremos, com a utilização do query()
dados.query('composto in @comp_index')
```

```python
# plotar um gráfico com a biblioteca Seaborn e ajustes visuais com matplot
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'composto',data = dados.query('composto in @comp_index'))
ax.set_title('Top 5 Compostos')
plt.show()
```

### Desafio 01 - Ordenar o gráfico countplot.

```python
sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'composto',data = dados.query('composto in @comp_index'), order = comp_index)
ax.set_title('Top 5 Compostos')
plt.show()
```

### Desafio 02 - Melhorar a visualização alterando o tamanho da fonte e etc.

```python
sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @comp_index'), order = comp_index, lw = 4, ec = 'black', hatch = '/')

ax.set_title('Top 5 Compostos', fontsize = 20)
ax.set_xlabel('Compostos', fontsize = 15)
ax.set_ylabel('Total Experimentos', fontsize = 15)

plt.show()
```

### Desafio 03 - Plotar os histogramas com seaborn.

```python
dados_filtrados = dados.query('composto in @comp_index')

sns.set()
ax = sns.histplot(x = 'composto', data = dados_filtrados, kde = True)
plt.show()
```

### Desafio 04 - Estudar sobre as estatísticas retornadas no .describe().

```python
dados.describe()
```

- count: É a contagem de quantas linhas existem no DataFrame;

- mean: Média dos dados, de acordo com as colunas;

- std: Desvio padrão. Raiz quadrada da variância, que mede a disperção dos dados em todo da média;

- min: Valor mínimo encontrado;

- 25%: 1º quartil;

- 50%: Mediana ou 2º quartil;

- 75%: 3º quartil;

- max: Valor máximo encontrado.

### Desafio 05 - Refletir sobre a manipulação do tamanho das visualizações.

Aumentar o tamanho das visualizações podem molhorar o entendimento das informações que estão sendo representadas pelo grágico, mas o aumento do tamanho, pode gerar algumas distorções visuais nas informações. Por isso é necessário se atentar ao tipo de informações que você deseja transmitir com os gráficos e utilizar o mais adequado para a visualização.
